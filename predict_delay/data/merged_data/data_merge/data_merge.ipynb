```python
import pandas as pd

weather_fixed = pd.read_csv('../../weather/additional_weather_data/weather_data_fixed.csv')
all_takeoff = pd.read_csv('../../takeoff/total_takeoff_data/2024_takeoff.csv')
```

```python
# 데이터 병합 (timestamp 기준으로 왼쪽 조인)
merged_df = all_takeoff.merge(weather_fixed, on='timestamp', how='left')

# 결과 확인
print(merged_df.head())
```

```python
merged_df.to_csv("./merged_data.csv", index=False)
```
